```python
# Complete the catAndMouse function below.
def catAndMouse(x, y, z):
    if abs(x-z) < abs(y-z): return "Cat A"
    elif abs(y-z) < abs(x-z): return "Cat B"
    else: return "Mouse C"
```

```python
if __name__ == '__main__':
    q = int(input())

    for q_itr in range(q):
        xyz = input().split()

        x = int(xyz[0])

        y = int(xyz[1])

        z = int(xyz[2])

        result = catAndMouse(x, y, z)

        print(result)
```
