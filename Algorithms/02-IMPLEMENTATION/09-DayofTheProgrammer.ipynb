```python
def dayOfProgrammer(year):
    # Write your code here
    leap_year = False
    
    if year == 1918:
        return f"26.09.{year}"
    
    if year % 4 == 0 and year > 1918:
        if (year % 100) != 0 or (year % 400) == 0:
            leap_year = True
        else:
            leap_year = False

    if year < 1918:
        if year % 4 == 0:
            leap_year = True
        else:
            leap_year = False 
            
    if leap_year == False:
        return f"13.09.{year}"
    else:
        return f"12.09.{year}"       
```
