## Apple and Orange

```python
import math
import os
import random
import re
import sys
#
# Complete the 'countApplesAndOranges' function below.
#
# The function accepts following parameters:
#  1. INTEGER s
#  2. INTEGER t
#  3. INTEGER a
#  4. INTEGER b
#  5. INTEGER_ARRAY apples
#  6. INTEGER_ARRAY oranges
```

```python
def countApplesAndOranges(s, t, a, b, apples, oranges):
    # Write your code here
    a_count = 0
    o_count = 0

    for i in apples:
        if s <= (a + i) <= t:
            a_count +=1
    for j in oranges:
        if s <= (b + j) <= t:
            o_count +=1

    print(a_count)
    print(o_count)  
```
