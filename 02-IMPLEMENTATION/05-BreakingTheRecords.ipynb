```python
def breakingRecords(scores):
    # Write your code here
    value_min = scores[0]
    value_max = scores[0]
    count_max = 0
    count_min = 0

    for i in range(len(scores)):
        if scores[i] > value_max:
            value_max = scores[i]
            count_max += 1

        if scores[i] < value_min:
            value_min = scores[i]
            count_min += 1

    return count_max, count_min
```
